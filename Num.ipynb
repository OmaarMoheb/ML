```python
from sklearn.metrics import r2_score
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.neighbors import KNeighborsRegressor
from sklearn.preprocessing import StandardScaler, PolynomialFeatures
from sklearn.ensemble import RandomForestRegressor
from sklearn import metrics

data = pd.read_csv('D:/MLE/data.csv')
print(data.head())
print(data.isnull().sum())
X = data[['bedrooms', 'bathrooms', 'sqft_living', 'sqft_lot']]
y = data['price']
X_train, X_test, Y_train, Y_test = train_test_split(X, y, test_size=0.2, random_state=1)

scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

poly = PolynomialFeatures(degree=2)
X_train_poly = poly.fit_transform(X_train_scaled)
X_test_poly = poly.transform(X_test_scaled)

lr = LinearRegression()
lr.fit(X_train_poly, Y_train)
predictions_lr = lr.predict(X_test_poly)
kn = KNeighborsRegressor()
kn.fit(X_train_scaled, Y_train)
predictions_kn = kn.predict(X_test_scaled)

rf = RandomForestRegressor()
rf.fit(X_train_scaled, Y_train)
predictions_rf = rf.predict(X_test_scaled)
plt.scatter(Y_test, predictions_lr)

plt.title('Linear Regression: Actual vs Predicted')
plt.xlabel('Actual Prices')
plt.ylabel('Predicted Prices')
plt.show()

plt.scatter(Y_test, predictions_kn)
plt.title('KNeighborsRegressor: Actual vs Predicted')
plt.xlabel('Actual Prices')
plt.ylabel('Predicted Prices')
plt.show()

print('Linear Regression Metrics:')
print('Mean Absolute Error:', metrics.mean_absolute_error(Y_test, predictions_lr))
print('Mean Squared Error:', metrics.mean_squared_error(Y_test, predictions_lr))
print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(Y_test, predictions_lr)))
print('R^2 Score:', r2_score(Y_test, predictions_lr))

print('\nKNeighborsRegressor Metrics:')
print('Mean Absolute Error:', metrics.mean_absolute_error(Y_test, predictions_kn))
print('Mean Squared Error:', metrics.mean_squared_error(Y_test, predictions_kn))
print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(Y_test, predictions_kn)))
print('R^2 Score:', r2_score(Y_test, predictions_kn))
```
